```python
def read_file(filepath):
    
    with open(filepath) as f:
        str_text = f.read()
    
    return str_text
```

```python
doc =read_file('shakespeare.txt')
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser', 'tagger', 'ner'])
```

```python
nlp.max_length = 1198623
```

```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
tokens = separate_punc(doc)
```

```python
# organize into sequences of tokens
train_len = 25+1 # 50 training words , then one target word

# Empty list of sequences
text_sequences = []

for i in range(train_len, len(tokens)):
    
    # Grab train_len# amount of characters
    seq = tokens[i-train_len:i]
    
    # Add to list of sequences
    text_sequences.append(seq)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
import tensorflow as tf

def create_model(vocabulary_size, seq_len):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Embedding(vocabulary_size, 25))
    model.add(tf.keras.layers.LSTM(150, return_sequences=True))
    model.add(tf.keras.layers.LSTM(150))
    model.add(tf.keras.layers.Dense(150, activation='relu'))
    model.add(tf.keras.layers.Dense(vocabulary_size, activation='softmax'))
    
    model.build((None, seq_len))  # Build the model specifying input shape
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model

```

```python
model = create_model(vocabulary_size+1, seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(X, y, batch_size=128, epochs=300,verbose=1)
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_lines=3, words_per_line=10):
    '''
    INPUTS:
    model : model that was trained on text data
    tokenizer : tokenizer that was fit on text data
    seq_len : length of training sequence
    seed_text : raw string text to serve as the seed
    num_lines : number of lines to be generated
    words_per_line : number of words per line to be generated by model
    '''
    
    # Final Output
    output_text = []
    
    # Create num_lines
    for _ in range(num_lines):
        # Initial Seed Sequence
        input_text = seed_text
        
        # Create words_per_line
        for _ in range(words_per_line):
            # Take the input text string and encode it to a sequence
            encoded_text = tokenizer.texts_to_sequences([input_text])[0]
            
            # Pad sequences to our trained rate 
            pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
            
            # Predict Class Probabilities for each word
            pred_probabilities = model.predict(pad_encoded, verbose=0)[0]
            
            # Sample the index of the next word based on predicted probabilities
            next_word_index = np.random.choice(len(pred_probabilities), p=pred_probabilities)
            
            # Convert index to word
            pred_word = tokenizer.index_word[next_word_index + 1]  # Adjust index by 1
            
            # Update the sequence of input text (shifting one over with the new word)
            input_text += ' ' + pred_word
            
            output_text.append(pred_word)
        
        # Add a newline after each line
        output_text.append('\n')
    
    # Make it look like multiple lines of text.
    return ' '.join(output_text)

```
